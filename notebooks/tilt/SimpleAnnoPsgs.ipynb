```python
def get_anno_now(ofn_anno,decision):
    dfnow=df_orig=df.query(f'decision=="{decision}"')
    if os.path.exists(ofn_anno):
        dfnow=pd.read_pickle(ofn_anno)
    print(f'# Changed so far ({decision}):',len(dfnow[dfnow.changed==True]) if 'changed' in dfnow.columns else 0)
    #if 'changed' in dfnow.columns:
    #    display(dfnow[dfnow.changed==True])
    dfnow=dfnow.drop_duplicates(keep='last')
    return dfnow
def get_anno_now_abs(): return get_anno_now(ofn_anno_abs,'Abstract')#.iloc[:3]
def get_anno_now_conc(): return get_anno_now(ofn_anno_conc,'Concrete')#.iloc[:3]
```

```python
def save_anno_df(dfnow,ofn):
    dfchanged=dfnow[dfnow.changed==True] if 'changed' in dfnow.columns else None
    print('# Rows changed:',len(dfchanged) if dfchanged is not None else 0)
    if dfchanged is not None:
        dfnow.to_pickle(ofn)

def anno_df(dfnow):
    dfnow=dfnow[dfnow.ifn_img.apply(lambda fn: os.path.exists(fn))]
    # just field periods for now
#     print(dfnow.shape)
    dfnow=dfnow[dfnow.period.isin({'C17','C18','C19'})]
#     print(dfnow.shape)
    dfnow=dfnow.fillna(False)
    return annotate(
        dfnow,
        example_column='ifn_img',
#         id_column='id_smpl',
        shuffle=True,
        task_type='multilabel-classification',
        buttons_in_a_row=4,
        options=labels,#_abs,
        display_fn=lambda fn: display(Image(fn)) if os.path.exists(fn) else display('Image not found'),
    )

def anno_abs():
    df_abs=get_anno_now_abs()
    return anno_df(df_abs)
def anno_conc():
    df_conc=get_anno_now_conc()
    return anno_df(df_conc)

def save_anno_abs():
    dfnow=get_anno_now_abs().append(res_abs)#.drop_duplicates('id_smpl',keep='last')
    return save_anno_df(dfnow,ofn_anno_abs)
def save_anno_conc():
    dfnow=get_anno_now_conc().append(res_conc)#.drop_duplicates('id_smpl',keep='last')
    return save_anno_df(dfnow,ofn_anno_conc)
```
