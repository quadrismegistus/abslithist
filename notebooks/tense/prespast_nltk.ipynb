```python
def get_text_pos(text, force=False, resume=True, lim_paras=10, shuffle_paras=False, progress=True, nlp_max_chars=3000):
    # # paths
    # opath = text.path_pos
    # odir = os.path.dirname(opath)
    # if not os.path.exists(odir):
    #     os.makedirs(odir)
    
    # get paragraphs
    o=[]
    if type(text)==str:
        paras = text.split('\n\n')
    else:
        paras = text.paras

    for para_i,para in enumerate(paras):
        para = ftfy.fix_text(para)
        sents = nltk.sent_tokenize(para)
        for sent_i,sent in enumerate(sents):
            toks = nltk.word_tokenize(sent)
            toks_tags = nltk.pos_tag(toks)

            for ti,(tok,tag) in enumerate(toks_tags):
                odx=dict(
                    para_i=para_i+1,
                    sent_i=sent_i+1,
                    word_i=ti+1,
                    word=tok,
                    pos=tag
                )
                o.append(odx)
    return pd.DataFrame(o)
```

```python
get_text_pos("He ran back to the store. He was going there. Life is a maze. Life presents us with options.")
```
