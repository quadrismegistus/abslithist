```python
# convert to prodigy

def to_prodigy(ofn,df_sample,force=False):
    # do not allow overwrite
    if not force and os.path.exists(ofn):
        print(f'{ofn} already exists!')
        return
    from bs4 import BeautifulSoup
    import html2markdown
    with open(ofn,'w') as of:
        for d in tqdm(df_sample.sample(frac=1).to_dict('records')):
            text=d['passage']
            dom=BeautifulSoup(text,'lxml')
            d2={'text':dom.get_text(), 'html':d['passage']}
            d2['meta']=dict((k,v) for k,v in d.items() if k!='passage')
            d2str=json.dumps(d2)
            of.write(d2str+'\n')            
```

```python
sample_ofn='../../data/psgs/data.psgs.CanonFiction.v4.jsonl'
```
