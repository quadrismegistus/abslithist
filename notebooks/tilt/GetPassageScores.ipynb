```python
def load_scores_bytext(ifn,rd={'_':' ', '|':'\n'}):
    if not os.path.exists(ifn): return
    df=pd.read_feather(ifn).set_index('i_tok')
    df['tok']=df['tok'].apply(lambda w: rd.get(w,w))
    df['tokl']=df['tokl'].apply(lambda w: rd.get(w,w))
    return df
    
# test
for t in C.texts():
    ifn=os.path.join(scoredir,t.corpus.name,t.id+'.ft')
    df=load_scores_bytext(ifn)
    if df is not None: break
df
```

```python
def divide_by_sent_windows(df,nmin=50):
    o=[]
    ipsg=0
    nnow=0
    df=df.sort_values(['i_para','i_sent'])
    for (para_i,sent_i),dfg in sorted(df.groupby(['i_para','i_sent'])):
        o+=[ipsg for n in range(len(dfg))]
        nnow+=dfg.is_recog.sum()        
        if nnow>=nmin:
            ipsg+=1
            nnow=0
    return o

# test
df['i_psg']=divide_by_sent_windows(df)
df
```

```python
def detokenize_psg(psgdf,valcol='val'):
    psgdf=psgdf.sort_values('i_tok')
#     txt=''.join(psgdf.tok)
    txt=' '.join(
        ''.join(sentdf.tok)
        for i,sentdf in sorted(psgdf.groupby('i_sent'))
    )
    data=odx={'txt':txt}
    words=[x for x in psgdf[psgdf.is_punct==0].tokl]
    recogwords=[x for x in psgdf[psgdf.is_recog==1].tokl]
    
    data['i_tok']=min(psgdf.index)
    for icol in ['i_para','i_sent','i_word']:
        data[icol]=min(psgdf[icol])

    data['num_word']=len(words)
    data['num_word_types']=len(set(words))
    data['ttr']=data['num_word_types'] / data['num_word'] if data['num_word'] else np.nan

    data['num_recog']=len(recogwords)
    data['num_recog_types']=len(set(recogwords))
    data['ttr_recog']=data['num_recog_types'] / data['num_recog'] if data['num_recog'] else np.nan


    data['val_mean']=psgdf['val'].mean()
    data['val_median']=psgdf['val'].median()
    data['val_stdev']=psgdf['val'].std()

    return pd.DataFrame([odx])
```

```python
def dopath(inp):
    try:
        path,ofn=inp

        # load scores for text
        df=load_scores_bytext(path)

        # add psg boundaries
        df['i_psg']=divide_by_sent_windows(df)

        # summarize by psg
        dfpsgs=pmap_groups(
            detokenize_psg,
            df.groupby('i_psg'),
            num_proc=1,
            progress=False,
        )

        # save
        odir=os.path.dirname(ofn)
        if not os.path.exists(odir): os.makedirs(odir)
        dfpsgs.to_csv(ofn)    
        del dfpsgs
    #     return dfpsgs
    except Exception:
        pass
```
