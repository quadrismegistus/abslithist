```python
dfscores = get_allnorms()
scored = dict((x,y) for x,y in zip(dfscores.index, dfscores['Abs-Conc.Median.C18']) if safebool(x) and safebool(y))
# scored
```

```python
old=[]
for wd in words_ld:
    for wi,w in enumerate(wd['words']):
        odx={**wd, **dict(word=w, rank=wi+1, concreteness=scored.get(w,np.nan))}
        del odx['words']
        old.append(odx)
odf=pd.DataFrame(old)
odf['period'] = [f'{x} – {y}' for x,y in zip(odf.period_pos, odf.period_neg)]
odf['focus_period'] = [f'{prd} ({foc})' for prd,foc in zip(odf.period, odf.focus)]
odf['rank_word']=[f'{word} ({rank})' for rank,word in zip(odf['rank'], odf['word'])]
odf
```

```python
odf['']=''
odf['focus_period']=pd.Categorical(odf.focus_period, categories=[
    '1720-1755 – 1756-1791 (bubble)',
    '1720-1755 – 1756-1791 (burst)',
    '1756-1791 – 1720-1755 (bubble)',
    '1756-1791 – 1720-1755 (burst)'
])
p9.options.figure_size=(8,6)
fig=p9.ggplot(odf,p9.aes(x='', y='concreteness', color='period',label='word'))
fig+=p9.scale_color_brewer(palette=2, type='qual')
fig+=p9.geom_violin(draw_quantiles=[0.5])
fig+=p9.geom_text(data=odf.dropna().groupby('focus_period').head(10), angle=45, size=8)
fig+=p9.coord_flip()
fig+=p9.facet_wrap('focus_period')
fig+=p9.theme_minimal()
fig+=p9.labs(
    title='Distribution of concreteness scores across closest words to "bubble" and "burst"',
    y="Concreteness score (C18 variant)"

)
fig+=p9.theme(
    plot_title=p9.element_text(size=9, weight='bold'), 
    legend_position='none',
    text=p9.element_text(size=9)
)
fig.save('bubble_burst_word_scores.png')
```
