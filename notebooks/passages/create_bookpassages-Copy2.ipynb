```python
# gen_bookpassages??
def gen_bookpassages(t_or_idx,corpus=None,fname=None,save=False,periods={'median'},sources={'Median'},modernize=False):
    import lltk
    if type(t_or_idx)==str:
        if not corpus: return
        C=lltk.load(corpus) if type(corpus)==str else corpus
        if not idx in C.textd: return
    else:
        t=t_or_idx

    ld=count_absconc_path(
        t.path_txt,
        sources=sources,
        periods=periods,
        incl_psg=True,
        incl_eg=True,
        num_eg=25,
        psg_as_markdown=True,
        markdown_uses_html=False,
        modernize=modernize,
        progress=True
    )
    df=pd.DataFrame(ld)
    df['abs-conc']=df['num_abs']-df['num_conc']
    df['id']=t.id
    df=df.merge(t.corpus.metadata,on='id',how='left')
    if save: save_bookpassages(df,fname=fname if fname else t.id, incl_stats=True)
    return df
```

```python
df=gen_bookpassages(
    C.textd['Richardson.Pamela'],
    fname='pamela1',
    save=True,
    modernize=False,
    periods={'C18'},
    sources={'Median'}
)
df.median()
```
