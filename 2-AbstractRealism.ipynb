## Analysis

```python
dfall=pd.read_csv('data/counts/data.absconc.CanonFiction.csv').dropna()
dfall['abs/conc']=dfall['num_abs']/dfall['num_conc']
dfall['abs-conc']=dfall['num_abs']-dfall['num_conc']
df=dfall[dfall.num_total==max(dfall.num_total)].sort_values('abs/conc')
df
```

```python
dftext=df[df.period!='orig'].groupby('id').median().sort_values('abs-conc')
dftext
```
