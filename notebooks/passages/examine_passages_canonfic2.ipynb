```python
import sys; sys.path.append('../..')
from abslithist.words import *
from abslithist.realism import *
```

```python
cname='CanonFiction'
# cdf=pd.read_feather(f'data/counts/data.absconc.{cname}.v6.csv.ft')
cdf=pd.DataFrame(readgen_jsonl('../../data/counts/data.absconc.CanonFiction.psgs.v9-zcut05.jsonl'))
cdf=cdf[cdf.num_total == cdf.num_total.max()]
cdf['abs-conc']=cdf['num_abs']-cdf['num_conc']
cdf['abs/conc']=cdf['num_abs']/cdf['num_conc']
# cdf
```

```python
import lltk
cdf['abs/conc']=cdf['num_abs']/cdf['num_conc']
cdf['abs-conc']=cdf['num_abs']-cdf['num_conc']
cdf['abs-conc_z']=zscore(cdf['abs-conc'])
dfall=cdf.merge(lltk.load('CanonFiction').metadata,on='id').sort_values('abs-conc')
dfall=dfall[
    ['id','year','title','author','passage','abs','conc','abs-conc','abs-conc_z','num_abs','num_conc','num_neither','num_types','num_total','num_tokens','neither']
]
dfall
```
