## Analysis

```python
dfall=pd.read_csv('data/counts/data.absconc.CanonFiction.v2.csv').dropna()
dfall['abs/conc']=dfall['num_abs']/dfall['num_conc']
df=dfall[dfall.num_total==max(dfall.num_total)].sort_values('abs/conc')
df
```

```python
df[df.id=='Julian of Norwich.Divine Love']
```

```python
dftext=df.groupby('id').median().sort_values('abs/conc')
dftext
```

```python
dftext.sort_values('num_types')
```

```python
df.sort_values('num_types')
```

```python
dftext.plot(x='num_abs',y='num_types',kind='scatter')
```
