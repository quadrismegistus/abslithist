```python
ifn='../../data/scores/data.canon_fic_scores.v2.ft'
```

```python
import lltk,pandas as pd,os
df=pd.read_feather(ifn).dropna()
C=lltk.load('CanonFiction')
alldf=df.groupby('id').mean().join(C.meta,rsuffix='_meta',how='inner').sort_values('mean')
alldf
```

```python
alldf.to_csv(os.path.splitext(ifn)[0] + '.csv')
```
